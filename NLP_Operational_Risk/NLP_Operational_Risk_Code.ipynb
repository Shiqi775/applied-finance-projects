**3. Create keyword lists:**

```python
# Initialize Lemmatizer
lemmatizer = WordNetLemmatizer()

keywords = {
    'attack': {
        'relevant': ['cyber-', 'cyber', 'networks', 'systems', 'products', 'services', 'datacenter', 'infrastructure'],
        'irrelevant': ['terror', 'war', 'contraband', 'bombs']
    },
    'threat': {
        'relevant': ['cyber-', 'cyber', 'networks', 'systems', 'products', 'services', 'datacenter', 'infrastructure'],
        'irrelevant': ['terror', 'simulator', 'disease', 'legal action', 'competitors']
    },
    'computer': {
        'relevant': ['malware', 'virus', 'viruses', 'intrusions'],
        'irrelevant': ['fires', 'product sales', 'warranty claim']
    },
    'information system': {
        'relevant': ['malware', 'virus', 'viruses', 'intrusions'],
        'irrelevant': ['fires', 'product sales', 'warranty claim']
    },
    'malicious': {
        'relevant': ['software', 'programs', 'third parties', 'attacks'],
        'irrelevant': []
    },
    'breaches': {
        'relevant': [],
        'irrelevant': ['fiduciary duty', 'fiduciary duties', 'covenant', 'credit', 'agreement']
    },
    'hacker': {'relevant': [], 'irrelevant': ['fiduciary', 'warranty', 'regulations', 'contract']},
    'hacking': {'relevant': [], 'irrelevant': ['fiduciary', 'warranty', 'regulations', 'contract']},
    'social engineering': {'relevant': [], 'irrelevant': ['fiduciary', 'warranty', 'regulations', 'contract']},
    'denial of service': {'relevant': [], 'irrelevant': ['fiduciary', 'warranty', 'regulations', 'contract']},
    'cyberattack': {'relevant': [], 'irrelevant': ['fiduciary', 'warranty', 'regulations', 'contract']},
    'cybersecurity': {'relevant': [], 'irrelevant': ['fiduciary', 'warranty', 'regulations', 'contract']}
}

```

**4. Implement the extraction logic:**

```python
import nltk

try:
    nltk.data.find('corpora/wordnet')
except LookupError:
    nltk.download('wordnet')
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    words = []
    for word in tokens:
        words.append(lemmatizer.lemmatize(word, 'v'))  # eg. deal with attacked, attacking
        words.append(lemmatizer.lemmatize(word, 'n'))  # eg. deal with attacker, attacks
    return words

def is_cybersecurity_related(sentence):
   tokens = preprocess_text(sentence)
   
   for keyword, conditions in keywords.items():
       if (lemmatizer.lemmatize(keyword, 'v') in tokens or 
           lemmatizer.lemmatize(keyword, 'n') in tokens):
           # Check irrelevant words
           if any(lemmatizer.lemmatize(irr, 'v') in tokens or 
                 lemmatizer.lemmatize(irr, 'n') in tokens
                 for irr in conditions['irrelevant']):
               return False
           
           # If no relevant words required
           if not conditions['relevant']:
               return True
               
           # Check relevant words
           if any(lemmatizer.lemmatize(rel, 'v') in tokens or 
                 lemmatizer.lemmatize(rel, 'n') in tokens
                 for rel in conditions['relevant']):
               return True
   return False

# Apply the cybersecurity-related extraction function to text data
text_df['cybersecurity_sentences'] = text_df['sentences'].apply(
    lambda x: [s for s in x if is_cybersecurity_related(s)]
)
text_df['cybersecurity_sentence_count'] = text_df['cybersecurity_sentences'].apply(len)
text_df['total_sentence_count'] = text_df['sentences'].apply(len)
```

#### Step 5: Portfolio Analysis
---

- As all the companies in year 1996-1998 has 0 cybersecurity risk, it's not meaningful to cut portfolio into five quantile based on all 0 cybersecurity risk and make comparison in return for different quantile portfolio. Therefore, year 1996-1998 is dropped.
- For 1999-2010, given that over 80% of companies report zero cybersecurity risk, I use a binary classification: Quintile 1 for zero-risk companies and Quintile 5 for those with any non-zero risk.
- After 2010, as cybersecurity risk ratios become more diverse, I divide companies into five quintiles based on their cybersecurity risk levels and make comparison on equal-weighted and value-weighted return accordingly.

**1. Form equal weighted portfolios**

```python
# Equal-weighted portfolio returns
def portfolio_returns(data, weights='equal'):
    # Create a function to assign quintiles within each year
    def assign_year_quintiles(group):
        if (group['cybersecurity_risk'] == 0).all():  # Check if all values are zero
            return pd.DataFrame()  # Drop this year
        elif (group['cybersecurity_risk'] == 0).sum() > len(group) * 0.8:  # Check if most values are zero
            group['risk_quintile'] = group['cybersecurity_risk'].apply(lambda x: 1 if x == 0 else 5)
        else:
            # Manually calculate quintiles
            sorted_group = group.sort_values(by='cybersecurity_risk')
            n = len(sorted_group)
            quintile_size = n // 5
            labels = [1, 2, 3, 4, 5]
            bins = [0] * n
            for i in range(5):
                start = i * quintile_size
                end = (i + 1) * quintile_size if i < 4 else n
                for j in range(start, end):
                    bins[sorted_group.index.get_loc(sorted_group.iloc[j].name)] = labels[i]
            group['risk_quintile'] = bins
        return group

    # Apply the function to create quintiles for each year
    data = data.groupby('year').apply(assign_year_quintiles).reset_index(drop=True)
   
    if weights == 'equal':
        returns = data.groupby(['year', 'risk_quintile'])['1yr_future_RET'].mean().unstack()
    elif weights == 'value':
        data['weighted_return'] = data.groupby(['year', 'risk_quintile']).apply(
            lambda x: x['1yr_future_RET'] * x['MarketCap'] / x['MarketCap'].sum()
        ).reset_index(level=[0,1], drop=True)
        returns = data.groupby(['year', 'risk_quintile'])['weighted_return'].sum().unstack()
    
    returns = returns.reset_index()
    returns['year'] = returns['year'].astype(int)
    returns.set_index('year', inplace=True)
    
    # Calculate the difference between high and low risk portfolios
    returns['High Risk-Low Risk'] = returns[5] - returns[1]
    
    # Calculate the average returns for quintile 5 and quintile 1 across all years
    avg_quintile_5_return = returns[5].mean()
    avg_quintile_1_return = returns[1].mean()
    
    return returns, avg_quintile_5_return, avg_quintile_1_return
```

```python
equal_weighted, avg_quintile_5_return, avg_quintile_1_return = portfolio_returns(sample_df, weights='equal')

# Print the average returns for quintile 5 and quintile 1
print(f"\nAverage return for quintile 5 across all years: {avg_quintile_5_return:.4f}")
print(f"Average return for quintile 1 across all years: {avg_quintile_1_return:.4f}")
```
